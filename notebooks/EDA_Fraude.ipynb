```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style('whitegrid')
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
df= pd.read_csv('../data/creditcard.csv')
```

```python
print("Apercu des donnees :")
display(df.head())

print(f"\n Dimensions du jeu de donnees: {df.shape[0]} lignes et {df.shape[1]} colonnes")

print("\nInformations sur les colonnes:")
df.info()

print("\n Resume statistique :")
display(df.describe())
```

```python
class_counts=df['Class'].value_counts()
print("Distribution des classes :")
print(class_counts)

plt.figure(figsize=(8,6))
sns.countplot(x='Class', data=df)
plt.title('Distribution des Transactions: Normales (0) Vs Fraudes (1)')
plt.xlabel('Classe')
plt.ylabel('Nombre de transactions')
plt.show()

fraud_percentage= (class_counts[1]/ class_counts.sum())*100
print(f"\nPourcentage de transactions frauduleuses: {fraud_percentage:.4f}%")
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df['Amount'], bins=100, kde=True)
plt.title('Distribution des Montants des Transactions')
plt.xlabel('Montant (€)')
plt.ylabel('Frequence')
plt.xlim(0,1000)
plt.show()

print("\nStatistiques descriptives pour Amount:")
print(df['Amount'].describe())
```

```python
#Le temps est en secondes. Conversion en heures pour voir s'il y a un cycle journalier.
df['Hour']=df['Time'].apply(lambda x:np.floor(x/3600))

plt.figure(figsize=(12,6))
sns.histplot(df['Hour'] % 24, bins=24)
plt.title('Distribution des Transactions au cours de la journee (Cycle de 24h)')
plt.xlabel('Heure de la journee')
plt.ylabel('Nombre de Transactions')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='Class', y='Amount', data=df)
plt.title('Montant des transactions par Classe')
plt.ylim(0,300)
plt.show()

print("\nStatistiques des montants pour les fraudes :")
display(df[df['Class'] == 1]['Amount'].describe())

print('\nStatistiques des montants pour les transactions normales :')
display(df[df['Class'] == 0]['Amount'].describe())
```

```python
from sklearn.preprocessing import StandardScaler

df_scaled=df.copy()

scaler=StandardScaler()

df_scaled[['Amount', 'Time']] =scaler.fit_transform(df_scaled[['Amount', 'Time']])

print("Apercu des donnees apres scaling de 'Time' et 'Amount': ")
display(df_scaled[['Time', 'Amount']].head())
```

```python
correlation_matrix=df_scaled.corr()

corr_with_class=correlation_matrix['Class'].sort_values(ascending=False)

print("Correlation des variables avec la classe de fraude :")
print(corr_with_class)

plt.figure(figsize=(20,15))
sns.heatmap(correlation_matrix, cmap='coolwarm_r', annot=False)

plt.title('Matrice de correlation de toutes les variables')
plt.show()
```

```python
positive_corr_vars= corr_with_class.head(3).index.drop('Class')
negative_corr_vars=corr_with_class.tail(2).index

fig, axes=plt.subplots(nrows=2, ncols=2, figsize=(18,10))
fig.suptitle('Distribution des variables les plus correlees a la fraude', fontsize=20)

for i, var in enumerate(positive_corr_vars):
    sns.kdeplot(data=df_scaled, x=var, hue='Class', ax=axes[0, i], fill=True, common_norm=False)
    axes[0, i].set_title(f'Distribution de {var}')


for i, var in enumerate(negative_corr_vars):
    sns.kdeplot(data=df_scaled, x=var, hue='Class', ax=axes[1, i], fill=True, common_norm=False)
    axes[1, i].set_title(f'Distribution de {var}')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
